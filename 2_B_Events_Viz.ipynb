# Data Viz

## Step 1: Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
from datetime import datetime, date
```

## Step 2: Reading the data

```python
df = pd.read_csv('/Users/ksah4444/Documents/Capstone_UFC_ML_KAGGLE/Data/df_events.csv')
```

```python
# Drop the 'Unnamed: 0' column
df.drop(['Unnamed: 0'], axis = 1, inplace=True)
```

```python
# First, let's seperate the month, day, and year to calculate the year of the fight
df[["Month", "Day", "Year"]] = df["date"].str.split(" ", expand=True)
df
```

```python
# Let's look at how the ufc events over the years are distributed
df['Year'].value_counts().plot(kind='barh', figsize=(8, 6))
plt.xlabel("number of events", labelpad=14)
plt.ylabel("year", labelpad=14)
plt.title("Events per year", y=1.02);
```

```python
# Is there more popular months than other?
df['Month'].value_counts().plot(kind='barh', figsize=(8, 6))
plt.xlabel("number of events", labelpad=14)
plt.ylabel("Month", labelpad=14)
plt.title("Events per month", y=1.02);
```

```python
# Failed attempt to plot the location
df['location'].value_counts().plot(kind='barh', figsize=(8, 6))
plt.xlabel("number of events", labelpad=14)
plt.ylabel("location", labelpad=14)
plt.title("Events per location", y=1.02);
```

```python
# How long does a ufc fight last?
df['last_round'].value_counts().plot(kind='barh', figsize=(8, 6))
plt.xlabel("number of fights", labelpad=14)
plt.ylabel("Number of rounds", labelpad=14)
plt.title("Distribution of fight length", y=1.02);
```
