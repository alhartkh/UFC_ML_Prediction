## Step 4: Clean df_events

```python
# Drop the columns we don't need
df_events1 = df_events[["R_fighter","B_fighter","win_by","last_round","date","location","Fight_type","Winner"]].copy()
```

```python
# Check if there is fights with no winner
null_name = pd.isnull(df_events1["Winner"])
df_events1[null_name]
```

```python
# Drop fights with no winner
df_events2 = df_events1.dropna(subset=['Winner'])
```

```python
# Let's see what are the values in the "win_by" column
df_events2["win_by"].value_counts()
```

```python
# Let's drop the "DQ" values, althought it's a win.The outcomes vary a lot
df_events3 = df_events2[df_events2.win_by !="DQ"]
```

```python
# Check how many DQ wins is there
len(df_events2[df_events2.win_by =="DQ"])
```

```python
# Create a copy, to keep things clean
df_events4 = df_events3
```

```python
# We want to red side to be all the winners, to make it easier to split later
idx =  np.where(df_events4['Winner'] == df_events4['B_fighter'], True, False)
df_events4.loc[idx] = df_events4.loc[idx].rename(columns={'B_fighter':'R_fighter','R_fighter':'B_fighter'})
```

#### https://stackoverflow.com/questions/25792619/what-is-correct-syntax-to-swap-column-values-for-selected-rows-in-a-pandas-data

```python
# Check if redside is all winners
df_events4["R_fighter"].equals(df_events4["Winner"])
```

```python
# Split the winners equally 
swap_indices = np.random.choice(len(df_events4), size = len(df_events4) // 2, replace = False)
df_events4.iloc[swap_indices, [0, 1]] = df_events4.iloc[swap_indices, [1, 0]].values
```

```python
# Print the length of the column to make sure the split was done right
print(len(df_events4))
print(len(df_events4)/2)
```

```python
# Check if there was an even split
df_events4["Winner"] = df_events4["Winner"] == df_events4["R_fighter"]
df_events4["Winner"] = df_events4["Winner"].astype(int)
df_events4["Winner"].value_counts()
```

```python
# Check if it was done right
df_events4.head()
```

```python
# Let's handle the fighters that were duplicates in the df_fighters
df_events5 = df_events4.copy()
for col in ["R_fighter", "B_fighter"]:
    df_events5.loc[(df_events5[col] == "Michael McDonald") &
                    (df_events5["Fight_type"] == "Light Heavyweight Bout"), col] = "Michael McDonald 205"
    
    df_events5.loc[(df_events5[col] == "Bruno Silva") &
                    (df_events5["Fight_type"] == "Flyweight Bout"), col]= "Bruno Silva 125"
    
    df_events5.loc[(df_events5[col] == "Joey Gomez ") &
                    (df_events5["Fight_type"] == "Bantamweight Bout"), col] = "Joey Gomez 135"
    
    df_events5.loc[(df_events5[col] == "Tony Johnson") &
                    (df_events5["Fight_type"] == "Heavyweight Bout"), col] = "Tony Johnson 265"
```

```python
# Check if all fighters are in the figher database
fighter_names = df_fighters.index.values.tolist()

df_events6 = df_events5.loc[(df_events5["R_fighter"].isin(fighter_names)) &
                              (df_events5["B_fighter"].isin(fighter_names))]
```

```python
# Reset Index
df_events6.reset_index(inplace=True, drop=True)
```

```python
# Check how many fights were removed
print(len(df_events))
print(len(df_events6))
```

### Combining 

```python
# Create a df with only red side fighters
# Create a df with only blue side fighters
df_R = df_fighters.loc[df_events6["R_fighter"]]
df_R = df_R.add_suffix('fighter1')
df_B = df_fighters.loc[df_events6["B_fighter"]]
df_B = df_B.add_suffix('fighter2')
```

```python
# Combine the two df's together
df_R.reset_index(inplace=True, drop=True)
df_B.reset_index(inplace=True, drop=True)
df_combined = pd.concat([df_events6, df_R, df_B], axis=1, sort=False)
```
